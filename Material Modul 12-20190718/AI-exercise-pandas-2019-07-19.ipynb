```python
import pandas as pd
```

```python
df = pd.read_csv('flights.csv')
df
```

Berapa banyak penerbangan di bulan Januari?

```python
len(df[df['month'] == 1])
```

Berapa banyak penerbangan di bulan Januari hingga April?

```python
len(df[df['month'] < 5])
```

Berapa banyak penerbangan di bulan Januari yang berasal dari bandara JFK?

```python
len(df[(df['month'] == 1) & (df['origin'] == 'JFK')])
```

Berikan _tail number_, tujuan dan _delay_ kedatangan untuk penerbangan bulan Januari dari JFK yang tidak mengalami _delay_ keberangkatan.

```python
df[(df['month'] == 1) & (df['origin'] == 'JFK') & (df['dep_delay'] <= 0)][['tailnum', 'dest', 'arr_delay']]
```

Berikan _tail number_ dari semua penerbangan US Airways (kode maskapai: US). Ada berapa _tail number_ unik dari penerbangan-penerbangan US Airways?

```python
dfusairways = df[df['carrier'] == 'US']
print(dfusairways['tailnum'].dropna().unique(), "\n", len(dfusairways['tailnum'].dropna().unique()))
```

```python
print(df[df['carrier'] == 'US']['tailnum'].dropna().unique(), "\n", len(df[df['carrier'] == 'US']['tailnum'].dropna().unique()))
```

Tampilkan penerbangan dengan _delay_ keberangkatan terlama di bulan Januari.

```python
dfjan = df[df['month'] == 1]
dfsortedjanuary = dfjan.sort_values(by = 'dep_delay')
dfsortedjanuary.dropna().tail(1)
```

```python
d = df[df['month'] == 1]
dep = d['dep_delay'].agg('max') # nilai ini dipakai untuk memilih penerbangan dengan dep_delay terlama
df[df['dep_delay'] == dep]
```

```python
d = df[df['month'] == 1]
d[d['dep_delay'] == d['dep_delay'].max()]
```

Tunjukkan asal dan tujuan penerbangan selain milik maskapai US, UA, DL, dan B6.

```python
df[(df['carrier'] != 'US') & (df['carrier'] != 'UA') & (df['carrier'] != 'DL') & (df['carrier'] != 'B6')][['origin', 'dest']]
```

```python
df[~df['carrier'].isin(['US', 'UA', 'DL', 'B6'])][['origin', 'dest']] # dinegasikan
```

Berapa penerbangan yang terlaksana di setiap bulannya?

```python
df3 = df.sort_values(by = 'dep_time').dropna()

i = 1

for i in range(1,13,1):
    df3m = df3[df3['month' == i]].count()

print(df3m)
```

```python
df[df['dep_time'].notna()][['month', 'flight']].groupby(by = ['month']).count()
# menampilkan semua baris yang tiap elemen ada isinya, dapat dataframe yang berisi baris-baris yang memenuhi (tanpa nilai NaN)
```

Berapa banyak penerbangan setiap bulannya untuk setiap maskapai?

```python
df[df['dep_time'].notna()][['month', 'carrier', 'flight']].groupby(by = ['month', 'carrier']).count()
```

Berikan jumlah penerbangan setiap maskapai untuk setiap bulan jika jumlah tersebut melebihi 2000.

```python
grouped = df[df['dep_time'].notna()][['month', 'carrier', 'flight']].groupby(by = ['month', 'carrier']).count()
grouped = grouped.count() # hasil ekspresi ini adalah sebuah dataframe lagi
grouped[grouped['flight'] > 2000]
```
